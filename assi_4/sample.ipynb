```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import LabelEncoder , StandardScaler
data = pd.read_csv("bank_dataset.csv")
```

```python
# preprocessing
data = data.drop("id" , axis=1) #drop id
data["sex"] = LabelEncoder().fit_transform(data["sex"])
data["region"] = LabelEncoder().fit_transform(data["region"])
data["married"] = LabelEncoder().fit_transform(data["married"])
data["car"] = LabelEncoder().fit_transform(data["car"])
data["save_act"] = LabelEncoder().fit_transform(data["save_act"])
data["current_act"] = LabelEncoder().fit_transform(data["current_act"])
```

```python
#scaled data by z-score bec without it we had output all -1
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

#db scan
dbscan = DBSCAN(eps=1.2 , min_samples=3).fit(data_scaled)
print (dbscan.labels_)
```
