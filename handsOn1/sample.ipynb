```python
import pyfpgrowth
```

```python
import pandas as pd
df = pd.read_csv("store_data.csv")

# Convert rows to list of lists while removing NaN values
transactions = [
    [item for item in row if pd.notna(item)]
    for row in df.values
]
```

```python
FrequentPatterns=pyfpgrowth.find_frequent_patterns(transactions=transactions ,support_threshold=200)
print(FrequentPatterns)
Rules=pyfpgrowth.generate_association_rules(patterns=FrequentPatterns ,confidence_threshold= 0.2)
```
